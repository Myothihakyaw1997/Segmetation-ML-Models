## Preprocess Data

```python
df = pd.read_csv("imported_cars_2020-10-27 12_45_55.csv")
df['Brand Name'].replace('-',np.NaN,inplace=True)
df.dropna(subset = ["Brand Name"], inplace=True)
df
```

```python
df.head()
```

```python
y = df['Brand Name']
y
```

```python
x = df['Post Detail']
type(x)
```

## Changing ASCII code for x 

```python
import re
# x = x[0].split('\n')
ascii_x = []
for post in x:
    lines = post.split('\n')
    ascii_char= []
    for line in lines:
        for word in line:
            for char in word:
#                 ascii_x.append(ord(char.lower())/1000)
                ascii_char.append(ord(char.lower()))
#                 print(char,'a')
    ascii_x.append(ascii_char)
```

```python
df_x = pd.DataFrame(ascii_x)
```

```python
df_x
```

## FIll Null values with zero

```python
df_x.fillna(0,inplace=True)
```

```python
df_x
```

## Changing ASCII code for y

```python
import re
# x = x[0].split('\n')
ascii_y = []
for post in y:
    lines = post.split('\n')
    ascii_char= []
    for line in lines:
        for word in line:
            for char in word:
#                 ascii_x.append(ord(char.lower())/1000)
                ascii_char.append(ord(char.lower()))
#                 print(char,'a')
    ascii_y.append(ascii_char)
```

```python
df_y = pd.DataFrame(ascii_y)
```

## Fill Null values with zero 

```python
df_y.fillna(0,inplace=True)
df_y
```

## Training with RandomForest ML model 

```python
from sklearn.model_selection import train_test_split
```

```python
trainx,testx,trainy,testy = train_test_split(df_x,df_y,test_size=0.2)
testx.iloc[3]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rdf = RandomForestRegressor(n_estimators=100)
```

```python
train_rdf = rdf.fit(trainx,trainy)
```

```python
train_rdf.score(testx,testy)
```

```python
predx = np.array(testx.iloc[0])
predx = predx.reshape(len(predx))
predx.shape
```

## Predict the value of y by the input x
